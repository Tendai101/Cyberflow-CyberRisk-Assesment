```python

def assign_hacking_probability_score(network_architecture):
    # Initialize the default score
    probability_score = None

    # Use if statements to assign scores based on the network architecture
    if network_architecture == "Zero Trust Architecture (ZTA)" or network_architecture == "Microsegmented Network":
        probability_score = 1
    elif network_architecture == "Network Access Control (NAC)":
        probability_score = 2
    elif network_architecture == "Distributed Networks" or network_architecture == "Complex Network Topologies":
        probability_score = 3
    elif network_architecture == "Flat Perimeter Networks":
        probability_score = 4
    elif network_architecture == "Legacy Systems" or network_architecture == "Bring Your Own Device Networks" or network_architecture == "None":
        probability_score = 5
    else:
        return "Network architecture not recognized. Please enter a valid network architecture."

    return probability_score

# Get the network architecture from the user
network_architecture = input("Enter the company's Network Architecture: ")

# Call the function to assign the hacking probability score
probability_score_res = assign_hacking_probability_score(network_architecture)

# Print the assigned hacking probability score
print(probability_score_res )



```

```python
def assign_impact_score(industry):
    # Initialize the impact score
    impact_score = 0

    # Assign impact scores based on the industry
    if industry == "Finance" or industry == "Banking" or industry == "Manufacturing":
        impact_score = 16
    elif industry == "Healthcare" or industry == "Pharmaceuticals":
        impact_score = 8
    elif industry == "Technology" or industry == "Software":
        impact_score = 4
    elif industry == "Retail" or industry == "E-commerce":
        impact_score = 2
    else:
        impact_score = 1

    return impact_score

# Input the company's industry
company_industry = input("Enter the company's industry: ")

# Get and display the impact score
impact = assign_impact_score(company_industry)
print(f"The impact score for the {company_industry} industry is: {impact}")
```

```python
def calculate_average_rating(selected_animals):
    total_ratings = 0
    total_animals = 0

    for animal, rating in selected_animals.items():
        if rating >= 1 and rating <= 5:  # Check if rating is within range
            total_ratings += rating
            total_animals += 1

    if total_animals == 0:
        return "No animals selected."
    else:
        average_rating = total_ratings // total_animals  # Calculate average as an integer
        return f"Average rating: {average_rating}"

# Example usage:
animals = {
    "Dog": 4,
    "Cat": 5,
    "Elephant": 3,
    "Lion": 2,
    "Tiger": 4
}

print(calculate_average_rating(animals))
```
